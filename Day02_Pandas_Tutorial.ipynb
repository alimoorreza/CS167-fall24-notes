## Creating DataFrame from a 1D List

```python
# Example#2: initializing a DataFrame with list of items (without a column name)
data = ["reza", "chris", "eric"]
df_1 = pd.DataFrame(data)
print(df_1)
```

```python
# Example#3: initializing a DataFrame with list of items (without a column name)

data_list = [10, 20, 30, 40, 50, 60] # initialize list elements

df_1 = pd.DataFrame(data_list, columns=['numbers']) # Create the pandas DataFrame with column name is provided explicitly
print('size of the dataframe df_2', df_1.shape)
# print dataframe
df_1
```

```python
# Example#4: adding a column name (ie, "last name") to the DataFrame
df_1 = pd.DataFrame(data, columns=["last name"])
print(df_1)
```

## Creating DataFrame from a Dictionary

```python
# Example#5: Create the pandas DataFrame with the column names provided explicitly
data_dict = {'col1':[1,2,3], 'col2':[4,5,6], 'col3':[7,8,9]}
df_2 = pd.DataFrame(data_dict)
print('size of the dataframe df_2', df_2.shape)
# print dataframe
df_2
```

```python
# Example#6: Initializing a DataFrame with a dictionary of items allows you to specify the column names along with their corresponding values.
data_source = {'first name': ['a', 'b', 'c'], 'last name':['A', 'B', 'C']}
df_2 = pd.DataFrame(data_source)
print(df_2)

# Example#7: Initializing a DataFrame with a dictionary of items allows you to specify the column names along with their corresponding values.
data_source = {"first name":["alimoor", "chris", "eric"], "last name":["reza", "porter", "manley"], "scores":[2, 3, 4]}
df_2 = pd.DataFrame(data_source)
df_2.head()
```

## Selecting Rows in DataFrames using `loc` and `iloc`:

Why might we want a subset of the columns of a DataFrame?

Sometimes you don't need all of the columns and just want to work with a __subset__ of the columns of the original dataset. Other times, you may want to reorder the columns in your dataset.

Here's how you would do either of those:
The syntax for subsetting columns from a DataFrame (`df`) is:
- One column: `df['column_name']`
- Multiple columns: `df[['column1', 'column2', 'target']]`

So, if we wanted to look at the `price` column, we could do:

```python
import pandas as pd
path = '/content/drive/MyDrive/cs167_fall24/datasets/restaurant.csv'
restaurant_data = pd.read_csv(path)
print('data is a ', type(restaurant_data))
restaurant_data.head()
```

```python
prices = restaurant_data['price']
prices
```

```python
#prices = restaurant_data[[ 'est','price','target']]
prices = restaurant_data[[ 'alt','fri','pat']]
prices
```

Imagine you want to only work with `'rain'`, `'hun'`, and `'target'`:

```python
col_subset = restaurant_data[['rain', 'hun', 'target']]
col_subset.head()
```

Maybe you want to re-order your new subset so that `rain` and `hun` are switched:

```python
reordered = col_subset[['hun', 'rain', 'target']]
reordered.head()
```

## Syntax to Subset Rows only:

Usually when subsetting rows, we're using **conditions**.

Remember: when we apply a condition to a DataFrame, the result is a Series of Boolean values:

```python
restaurant_data.type == 'French'
```

```python
# save the condition as a variable (of a series of True/False)
condition = restaurant_data['target'] == 'No'
subset_rows = restaurant_data[condition]
subset_rows.head()
```

```python
# these two step process can be done in a single line
# condition = restaurant_data['target'] == 'No'
subset_rows = restaurant_data[restaurant_data['target'] == 'No']
subset_rows.head()
```

```python
# another example
italian_restaurants = restaurant_data[restaurant_data['type'] == 'Italian']
italian_restaurants.head()
```
